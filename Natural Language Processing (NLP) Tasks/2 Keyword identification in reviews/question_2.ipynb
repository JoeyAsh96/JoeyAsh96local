```python
from nltk import word_tokenize, pos_tag, bigrams, regexp_tokenize, corpus
```

```python
file = open(file='/content/drive/MyDrive/MSc Data Science & Artificial Intelligence/2nd Semester/CIS4515 Practical Data Analysis/CW 1/Task 2/positive_reviews_of_Video_Games.txt', mode = 'r')
lines = file.readlines()
doccie = ''
for line in lines:
    doccie += line.split('\t')[2].lower()
file.close()

stop_words = set(corpus.stopwords.words('english'))

wds = regexp_tokenize(doccie, pattern = r"[\w']+")

wds = [wrd for wrd in wds if not wrd in stop_words]

wds_with_pos = pos_tag(wds)

bigrams_pos = list(bigrams(wds_with_pos))
```

- A frequency calculation function called ***freq_of_bigrams()*** was created.
- The function takes pos_tagged bigrams as input and gives an output of each bigram and its frequency as observed from the tokenized word list.

```python
def freq_of_bigrams(bigrams_pos):
#An empty dictionary is used to instantiate the repository of bigrams and their frequencies.
    bigrams_freqs = {}

    for frst_e, snd_e in bigrams_pos:
        frst_w, frst_pos = frst_e[0], frst_e[1]
        snd_w, snd_pos = snd_e[0], snd_e[1]
#An 'if' statement is added evaluate if the bigram has been ecountered before in the
#the loop. If true, the counter adds 1 to the frequency value for the bigram.
        if (frst_w, snd_w) in bigrams_freqs:
            bigrams_freqs[(frst_w, snd_w)] += 1
#If the loop is encountering the bigram for the first time, it adds the bigram and
#assigns a value of 1 to it.
        else:
            bigrams_freqs[(frst_w, snd_w)] = 1
    return bigrams_freqs
```

```python
freq_of_bigrams(bigrams_pos)
```

- The frequency list of the bigrams is then sorted in descending order using the code below:

```python
#The code sorts the dictionary produced by the freq_of_bigrams function by descending order of the values.
#The top 40 bigrams by frequency are given below.
sorted_freq_of_bigrams = dict(sorted(freq_of_bigrams(bigrams_pos).items(), key=lambda item: item[1], reverse=True)[:40])
```
